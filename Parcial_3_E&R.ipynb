####Modelo de regresión lineal simple

```python
libros_data = pd.read_csv("data_libros.csv")
libros_data.head()
```

```python
libros_data["num_reviews"]=libros_data["num_reviews"].apply(str).str.replace(',', '') #convertir de string a float 
libros_data
```

```python
libros_data2 = libros_data[["rating", "num_reviews"]].astype(float)
libros_data2 = libros_data2.head(50)
libros_data2
```

```python
import matplotlib.pyplot as plt 
x = libros_data2["num_reviews"]
y = libros_data2["rating"]

plt.scatter(x, y)
plt.show()
```

###Gráfica de dispersión 
La gráfica consiste con el eje x siendo "num_reviews" y el eje y es "rating". Podemos ver que la mayoría de los libros se almacenan al principio de la gráfica entonces podemos decir que todos tienen un rango de rating y de reviews muy cercano. Hay algunos libros que son outlayers, esto significa que se separan de los demás libros por ejemplo tenemos un libro que tiene cerca de 140000 reviews y esto hace que se separe de los demás, tenemos otro ejemplo de unlibro con un rating de 4.75 siendo el lirbo con más rating

```python
#necesitamos la media de x y a la media de y 
import numpy as np
barx = np.mean(x)
bary = np.mean(y)
print(barx, bary)
```

```python
#debemos sacar la varianza
varx = np.var(x, ddof=-1) #variancia muestral #grados de libertad: cuanto le estas restando a la variable
covxy = np.cov(x, y, ddof= 1) #covarianza muestral 

varx, covxy
```

```python
print(varx)
```

```python
#quieres escoger solamente este numerito
covxy = covxy[0,1]
covxy
```

```python
#quien es beta 0
beta1 = covxy/varx 
beta0 = bary - beta1*barx
print(beta0, beta1)
```

##Fórmula Regresión Lineal 
y = 4.53 - 175.26

```python
#le agregamos a la gráfica de dispersión una línea de tendencia
import matplotlib.pyplot as plt
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x, y) 

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show
```

Podemos ver que la línea de tendencia es negativa. Ésta nos ayuda a ver la tendencia con mayor clarida que tiene nuestra gráfica

###Explicación del modelo 
En este modelo de regresión simple podemos ver que es un mal modelo ya que r cuadrada es 0.005 y como podemos ver está muy lejos de 1 esto significa que mientras más lejos del 1 peor modelo es. 

####Modelo de regresión lineal múltiple

```python
import pandas as pd
from statsmodels.formula.api import ols
```

```python
sales=list(libros["sales"]) #lista de sales
```

```python
libros2 = libros.head(100)
libros2
```

```python
Reg = ols(formula = "sales ~ rating + genre + name*rating", data = libros2) #regresión (~) respecto a gender e income, la primera es la variable responsiba (consumption)
Fit = Reg.fit() #alimentar modelo
Fit.summary()
#un modelo de regresión múltiple es con más coeficientes/variables
```

Tenemos tres variables en esta regresión lienal múltiple: tenemos rating, genre y name, con sales como variable responsiba. Podemos ver que nuestro modelo es bastante bueno ya que r cuadrada es 0.722. Al pricipio nos salía una r cuadrada muy rara pero mientras más variables le agregabamos mejor modelo salía. Tuvimos que recortar el data frame a 100 variables ya que sino eran muchas y no era un modelo confiable
